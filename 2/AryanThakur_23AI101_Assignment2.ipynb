```python
from pathlib import Path

# Directory containing text files
text_dir = Path('/kaggle/input/ai-governance-documents-data/agora/fulltext/')

# Get all .txt files
text_files = list(text_dir.glob('*.txt'))

# Print properties of each text file
for file in text_files:
    # File size
    file_size = file.stat().st_size  # Size in bytes
    # Read the content to check number of lines and characters
    with open(file, 'r') as f:
        content = f.read()
        num_lines = len(content.split('\n'))
        num_chars = len(content)
    
    print(f"Properties of {file.name}:")
    print(f"Size: {file_size / 1024:.2f} KB")
    print(f"Number of lines: {num_lines}")
    print(f"Number of characters: {num_chars}")
    print('-' * 40)
```

```python
import pandas as pd
import numpy as np

# Read the documents CSV file (or any other CSV file you'd like to work with)
documents_df = pd.read_csv('/kaggle/input/ai-governance-documents-data/agora/documents.csv')

# Let's assume we only want the numeric columns for matrix operations
numeric_df = documents_df.select_dtypes(include=[np.number])

# Step 1: Transpose of the matrix
transposed_matrix = numeric_df.T  # Transpose using pandas

# Step 2: Inversion of the matrix (only possible if it's square and invertible)
if numeric_df.shape[0] == numeric_df.shape[1]:
    try:
        inverted_matrix = np.linalg.inv(numeric_df.values)  # Inverse using numpy
        print("Inverse of the matrix:")
        print(inverted_matrix)
    except np.linalg.LinAlgError:
        print("Matrix is not invertible.")
else:
    print("Matrix is not square, cannot compute inversion.")

# Step 3: Dot product of the matrix and its transpose
dot_product = np.dot(numeric_df.values, transposed_matrix.values)  # Dot product using numpy

# Displaying the results
print("\nTranspose of the matrix:")
print(transposed_matrix.head())  # Display the transpose (first 5 rows)

print("\nDot product of the matrix and its transpose:")
print(dot_product)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the CSV files (choose one file or multiple)
collections_df = pd.read_csv('/kaggle/input/ai-governance-documents-data/agora/collections.csv')
documents_df = pd.read_csv('/kaggle/input/ai-governance-documents-data/agora/documents.csv')
segments_df = pd.read_csv('/kaggle/input/ai-governance-documents-data/agora/segments.csv')
authorities_df = pd.read_csv('/kaggle/input/ai-governance-documents-data/agora/authorities.csv')

# Create a directory to save the plots
save_dir = '/kaggle/working/plots/'
import os
if not os.path.exists(save_dir):
    os.makedirs(save_dir)

# --- Line Graph (Example: Plotting a column over time or order) ---
def plot_line_graph(df, column):
    plt.figure(figsize=(10, 6))
    plt.plot(df[column], label=column)
    plt.title(f'Line Graph of {column}')
    plt.xlabel('Index')
    plt.ylabel(column)
    plt.legend()
    plt.grid(True)
    # Save to disk
    plt.savefig(save_dir + f'line_graph_{column}.png')
    plt.show()

# Example: Plotting 'some_column' from documents_df (replace with an actual numeric column name)
# plot_line_graph(documents_df, 'some_column')  # Uncomment and replace with actual column

# --- Bar Graph (Example: Categorical data count) ---
def plot_bar_graph(df, column):
    plt.figure(figsize=(10, 6))
    sns.countplot(x=column, data=df)
    plt.title(f'Bar Graph of {column}')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    # Save to disk
    plt.savefig(save_dir + f'bar_graph_{column}.png')
    plt.show()

# Example: Plotting 'some_category_column' (replace with an actual categorical column name)
# plot_bar_graph(documents_df, 'some_category_column')  # Uncomment and replace with actual column

# --- Histogram (Example: Distribution of a numerical column) ---
def plot_histogram(df, column):
    plt.figure(figsize=(10, 6))
    plt.hist(df[column], bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    # Save to disk
    plt.savefig(save_dir + f'histogram_{column}.png')
    plt.show()

# Example: Plotting the distribution of a numerical column like 'numeric_column'
# plot_histogram(documents_df, 'numeric_column')  # Uncomment and replace with actual column

# --- Pie Chart (Example: Proportions of a categorical column) ---
def plot_pie_chart(df, column):
    plt.figure(figsize=(8, 8))
    df[column].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=sns.color_palette('Set3', len(df[column].unique())))
    plt.title(f'Pie Chart of {column}')
    # Save to disk
    plt.savefig(save_dir + f'pie_chart_{column}.png')
    plt.show()

# Example: Plotting a pie chart of 'category_column' (replace with actual categorical column name)
# plot_pie_chart(documents_df, 'category_column')  # Uncomment and replace with actual column

# --- Scatter Plot (Example: Relationship between two numerical columns) ---
def plot_scatter_plot(df, column_x, column_y):
    plt.figure(figsize=(10, 6))
    plt.scatter(df[column_x], df[column_y], alpha=0.6)
    plt.title(f'Scatter Plot: {column_x} vs {column_y}')
    plt.xlabel(column_x)
    plt.ylabel(column_y)
    # Save to disk
    plt.savefig(save_dir + f'scatter_plot_{column_x}_vs_{column_y}.png')
    plt.show()

# Example: Plotting relationship between 'numeric_column1' and 'numeric_column2'
# plot_scatter_plot(documents_df, 'numeric_column1', 'numeric_column2')  # Uncomment and replace with actual columns
```
